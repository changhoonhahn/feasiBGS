```python
fig = plt.figure(figsize=(6,6))
sub = fig.add_subplot(111)
in_g15 = ((target['ra'] > 211.5) & (target['ra'] < 223.5) & 
         (target['dec'] > -2.) & (target['dec'] < 3.))
isstar = (target['type'] == 'PSF')
maglim = (UT.flux2mag(target['flux_r'], band='r') < 20.1)
sub.scatter(target['ra'][in_g15 & maglim & ~isstar], target['dec'][in_g15 & maglim & ~isstar], c='C0', s=30, label='Type != PSF')
sub.scatter(target['ra'][in_g15 & isstar], target['dec'][in_g15 & isstar], c='C1', s=30, label='Type = PSF')
sub.scatter(tile0['ra'], tile0['dec'], c='k', s=5, label='Tile')
sub.scatter(tile1['ra'], tile1['dec'], c='k', s=5)
sub.scatter(tile2['ra'], tile2['dec'], c='k', s=5)

sub.legend(loc='upper left', markerscale=2, handletextpad=0.1, frameon=True, fontsize=15)
sub.set_xlabel('RA', fontsize=25)
sub.set_xlim([212.7, 213])
sub.set_ylabel('Dec', fontsize=25)
sub.set_ylim([-1, -0.7])
```
