Explore the atmosphere class in `specsim.atmosphere.Atmosphere`

```python
import numpy as np
import desisim.simexp
import specsim.atmosphere
import astropy.units as u
# -- local -- 
import env
from feasibgs import util as UT
```

```python
w, flx = np.loadtxt(UT.dat_dir()+'sky/spec-sky.dat', unpack=True, skiprows=2, usecols=[0,1])
```

```python
config = desisim.simexp._specsim_config_for_wave((w[w < 9800.]*u.Angstrom).to('Angstrom').value, 
                                        dwave_out=None, specsim_config_file='desi')
atmosphere = specsim.atmosphere.initialize(config)
```

```python
atmosphere.plot()
```
