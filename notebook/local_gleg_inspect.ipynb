```python
import os
import numpy as np 
# -- local --
from feasibgs import util as UT
from feasibgs import catalogs as Cat
from feasibgs import _inspector as Inspect
```

```python
# read in GAMA-Legacy survey
cata = Cat.GamaLegacy()
gleg = cata.Read('g15')
redshift = gleg['gama-spec']['z']
```

```python
# read in the spectra of the 5th out of 64 block from the GAMA G15 field
f_spec = ''.join([UT.dat_dir(), 'spectra/' 
                  'GamaLegacy.g15.expSpectra.darksky.seed1.exptime300.5of64blocks.fits'])
f_zbest = ''.join([UT.dat_dir(), 'redrock/'
                   'GamaLegacy.g15.expSpectra.darksky.seed1.exptime300.5of64blocks.redrock.fits'])
f_zbest_arch = ''.join([UT.dat_dir(), 'redrock/'
                   'GamaLegacy.g15.expSpectra.darksky.seed1.exptime300.5of64blocks.redrock.archetypetest.fits'])
```

```python
f_index = ''.join([UT.dat_dir(), 'spectra/', 
                   'GamaLegacy.g15.expSpectra.darksky.seed1.exptime300.5of64blocks.index'])
index = np.loadtxt(f_index, unpack=True, usecols=[0], dtype='i')
```

```python
z_true = redshift[index]
```

```python
insp = Inspect.load_spectra(f_spec, zbestfile=f_zbest)
```

```python
zwarn = (insp.zbest['ZWARN'] != 0)
zfail = (np.abs(insp.zbest['Z'] - z_true)/(1.+z_true) > 0.003)
print('%i ZWARN != 0' % np.sum(zwarn))
print('%i redshift failures' % np.sum(zfail))
```
