Lets take a look at the tile centers that Kyle generated using the spacing suggested by David

```python
ftiles = fits.open(''.join([UT.dat_dir(), 'sv/bgs_sv.fits']))
tiles = ftiles[1].data
```

```python
fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(111)
for i in range(len(tiles['ra'])): 
    circi = mpl.patches.Circle((tiles['ra'][i], tiles['dec'][i]), 1.605, facecolor='None', edgecolor='C0', linewidth=1)
    sub.add_patch(circi)
rect_g15 = mpl.patches.Rectangle((211.5, -2.), 12, 5, facecolor='None', edgecolor='k', linewidth=2)
sub.add_patch(rect_g15)
sub.text(211.5, 3.5, 'G15', fontsize=20)
rect_g02 = mpl.patches.Rectangle((30.2, -6.), 8.6, 2, facecolor='None', edgecolor='k', linewidth=2)
sub.add_patch(rect_g02)
sub.text(30.2, -3.25, 'G02', fontsize=20)
rect_g09 = mpl.patches.Rectangle((129., -2.), 12, 5, facecolor='None', edgecolor='k', linewidth=2)
sub.add_patch(rect_g09)
sub.text(129., 3.5, 'G09', fontsize=20)
rect_g12 = mpl.patches.Rectangle((174, -3.), 12, 5, facecolor='None', edgecolor='k', linewidth=2)
sub.add_patch(rect_g12)
sub.text(174., 2.5, 'G12', fontsize=20)
sub.set_xlabel('RA', fontsize=25)
sub.set_xlim([0., 280.])
sub.set_ylabel('Dec', fontsize=25)
sub.set_ylim([-12., 10.])
```

```python
fig = plt.figure(figsize=(6,6))
sub = fig.add_subplot(111)
for i in range(len(tiles['ra'])): 
    circi = mpl.patches.Circle((tiles['ra'][i], tiles['dec'][i]), 1.605, facecolor='None', edgecolor='C0', linewidth=1)
    sub.add_patch(circi)
rect_g15 = mpl.patches.Rectangle((211.5, -2.), 12, 5, facecolor='None', edgecolor='k', linewidth=2)
sub.add_patch(rect_g15)
sub.set_title('GAMA G15 Field', fontsize=30)
sub.set_xlabel('RA', fontsize=25)
sub.set_xlim([210., 225.])
sub.set_ylabel('Dec', fontsize=25)
sub.set_ylim([-7.5, 7.5])
```
