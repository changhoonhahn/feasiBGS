```python
sky_model = specsim.simulator.Simulator('desi').atmosphere
```

```python
def KS_sky(atbl): 
    sky_model.airmass = atbl['AIRMASS']
    sky_model.moon.moon_phase = np.arccos(2.*atbl['MOON_ILL'] - 1)/np.pi
    sky_model.moon.moon_zenith = (90. - atbl['MOON_ALT']) * u.deg
    sky_model.moon.separation_angle = atbl['MOON_SEP'] * u.deg
    return sky_model.surface_brightness
```

```python
ks_wlim = ((sky_model._wavelength.value > 4500.) & (sky_model._wavelength.value < 4700.))
def KS_sky460(atbl):
    sbright = KS_sky(atbl)
    return np.average(sbright.value[ks_wlim])
```

```python
ks460 = np.array([KS_sky460(boss_blue[i]) for i in range(len(boss_blue))])
```

```python
from feasibgs import catalogs as Cat
from feasibgs import forwardmodel as FM
```

```python
boss_red = aTable.read(''.join([UT.dat_dir(), 'sky/', 'Bright_BOSS_Sky_red.fits']))
```

```python
cata = Cat.GamaLegacy()
gleg = cata.Read('g15', dr_gama=3, dr_legacy=7)
redshift = gleg['gama-spec']['z'] # redshift
absmag_ugriz = cata.AbsMag(gleg, kcorr=0.1, H0=70, Om0=0.3, galext=False) # ABSMAG k-correct to z=0.1
i_rand = np.array([38164])

bgs3 = FM.BGStree()
match = bgs3._GamaLegacy(gleg, index=i_rand)
mabs_temp = bgs3.meta['SDSS_UGRIZ_ABSMAG_Z01']
s_bgs = FM.BGSsourceSpectra(wavemin=1500.0, wavemax=2e4)
vdisp = np.repeat(100.0, len(i_rand))

# r-band aperture magnitude from Legacy photometry
r_mag_gama = gleg['gama-photo']['modelmag_r'][i_rand]
r_mag_apflux = UT.flux2mag(gleg['legacy-photo']['apflux_r'][i_rand,1], method='log')
r_mag_legacy = UT.flux2mag(gleg['legacy-photo']['flux_r'][i_rand], method='log')

emline_flux = s_bgs.EmissionLineFlux(gleg, index=i_rand, dr_gama=3, silent=True)
flux_fm, wave_fm, _, _ = s_bgs.Spectra(r_mag_apflux, redshift[i_rand], vdisp, seed=1,        
                                     templateid=match, emflux=emline_flux, mag_em=r_mag_gama, silent=False)
```

```python
r_mag_gama, r_mag_apflux, r_mag_legacy
```

```python
w_blue = boss_blue['WAVE'][brightest]
w_red = boss_red['WAVE'][brightest]
w_boss_sky = np.concatenate([w_blue, w_red[w_red > w_blue.max()]]) * 10.
sky_boss = np.concatenate([boss_blue['SKY'][brightest], boss_red['SKY'][brightest][w_red > w_blue.max()]])/np.pi
```

```python
fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(111)
sub.plot(wave_fm, flux_fm[0], c='k')
sub.scatter(w_boss_sky, sky_boss, c='C0', s=1)
sub.scatter(sky_model._wavelength, KS_sky(boss_blue[brightest]), c='C1', s=1)
sub.set_xlim([3600., 9800.])
sub.set_ylim([0., 20.])
```

```python
fdesi = FM.fakeDESIspec()
```

```python
sbright_unit = 1e-17 * u.erg / (u.arcsec**2 * u.Angstrom * u.s * u.cm ** 2 )
```

```python
t_exp = 480.
bgs_boss = fdesi.simExposure(wave_fm, flux_fm, exptime=t_exp, 
                             skycondition={'name':'input', 'sky': sky_boss * sbright_unit, 'wave': w_boss_sky * u.Angstrom})
bgs_ks = fdesi.simExposure(wave_fm, flux_fm, exptime=t_exp, 
                           skycondition={'name':'input', 'sky': KS_sky(boss_blue[brightest]), 'wave':sky_model._wavelength})

fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(111)

snrs_boss, snrs_ks = [], [] 
for iband, band in enumerate(bgs_ks.bands): 
    sub.plot(bgs_boss.wave[band], bgs_boss.flux[band][0], c='C0', lw=1, label='BOSS Sky')
    sub.plot(bgs_ks.wave[band], bgs_ks.flux[band][0], c='C1', lw=1, label='UVES + KS')
    if band == 'b': sub.legend(loc='upper right', frameon=True, fontsize=15)
        
    # calculate SNR
    these = ((bgs_boss.wave[band] > np.mean(bgs_boss.wave[band])-50) &
             (bgs_boss.wave[band] < np.mean(bgs_boss.wave[band])+50) &
             (bgs_boss.flux[band][0] > 0))
    snr_boss = np.median(bgs_boss.flux[band][0, these] * np.sqrt(bgs_boss.ivar[band][0, these]))
             
    these = ((bgs_ks.wave[band] > np.mean(bgs_ks.wave[band])-50) &
             (bgs_ks.wave[band] < np.mean(bgs_ks.wave[band])+50) &
             (bgs_ks.flux[band][0] > 0))
    snr_ks = np.median(bgs_ks.flux[band][0, these] * np.sqrt(bgs_ks.ivar[band][0, these]))
    
    snrs_boss.append(snr_boss)
    snrs_ks.append(snr_ks)    
sub.plot(wave_fm, flux_fm[0], c='k')
sub.text(0.1, 0.9, r'$r_\mathrm{DECaLS} = '+str(round(r_mag_legacy[0],2))+'$', va='top', ha='left', transform=sub.transAxes, fontsize=20)
sub.set_xlim([3600., 9800.])
sub.set_ylim([0., 5.])
plt.show() 

print('SNR for b, r, z bands')
print('BOSS Sky', snrs_boss)
print('UVES + KS Sky', snrs_ks)
```
